```python
import geopandas as gpd
```

```python
# Load zone 4 polygon data from shape file

FILE_PATH_ZONE_4 = "/content/drive/MyDrive/IPAUA_Maz/shape_files/zone_4/layers/POLYGON.shp"

zone_4_polygon_data = gpd.read_file(FILE_PATH_ZONE_4)

# Extract the bounds of the polygon
minx, miny, maxx, maxy = zone_4_polygon_data.total_bounds

# Define zone 4 bounds using the extracted values
zone_4_bounds = {
    "latitude_min": miny,
    "latitude_max": maxy,
    "longitude_min": minx,
    "longitude_max": maxx,
}

print(f"Zone 4 Bounds:\n {zone_4_bounds}")
```
